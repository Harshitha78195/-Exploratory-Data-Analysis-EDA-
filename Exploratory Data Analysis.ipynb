```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("traffic_crash_data.csv")  # Replace with your actual dataset



```

```python
# 1. Crash Frequency by Hour
hour_counts = np.bincount(df['CRASH_HOUR'])
plt.figure(figsize=(10, 5))
plt.bar(range(len(hour_counts)), hour_counts, color='skyblue', edgecolor='black')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crashes')
plt.title('Crash Frequency by Hour')
plt.xticks(range(0, 24))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# **Insight**: Most crashes occur during rush hours (7-9 AM & 5-7 PM).
```

```python
# 2. Speed Limit vs. Injury Severity Distribution
plt.figure(figsize=(10, 5))
plt.scatter(df['POSTED_SPEED_LIMIT'], df['INJURIES_TOTAL'], alpha=0.5, color='red')
plt.xlabel('Posted Speed Limit (MPH)')
plt.ylabel('Total Injuries')
plt.title('Speed Limit vs. Injury Severity')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# **Insight**: Higher speed limits generally correlate with higher injury counts.
```

```python
# Calculate mean injuries per weather condition
weather_conditions = df["WEATHER_CONDITION"].unique()
avg_injuries = np.array([df[df["WEATHER_CONDITION"] == w]["INJURIES_TOTAL"].mean() for w in weather_conditions])

# Plot
plt.figure(figsize=(10,5))
plt.bar(weather_conditions, avg_injuries, color='green', alpha=0.7)
plt.xlabel("Weather Condition")
plt.ylabel("Average Number of Injuries")
plt.title("Weather Condition Impact on Crash Severity")
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#This grouped bar chart compares the average number of injuries under different weather conditions.

```

```python
# Aggregate the top 10 most frequent crash causes
top_causes = df["PRIM_CONTRIBUTORY_CAUSE"].value_counts().nlargest(10)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(top_causes.index, top_causes.values, color="c")
plt.xlabel("Primary Crash Causes")
plt.ylabel("Number of Crashes")
plt.title("Top 10 Most Common Crash Causes")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot histogram of posted speed limits
plt.figure(figsize=(10, 5))
plt.hist(df["POSTED_SPEED_LIMIT"], bins=20, color="g", edgecolor="black", alpha=0.7)
plt.xlabel("Speed Limit (mph)")
plt.ylabel("Crash Count")
plt.title("Distribution of Speed Limits in Crashes")
plt.grid()
plt.show()

#This histogram shows the frequency of crashes occurring at different posted speed limits.
```

```python
# Count occurrences of each injury severity
injury_counts = df["MOST_SEVERE_INJURY"].value_counts()

# Plot pie chart
plt.figure(figsize=(7, 7))
plt.pie(injury_counts, labels=injury_counts.index, autopct="%1.1f%%", colors=["lightcoral", "gold", "lightblue", "green"])
plt.title("Injury Severity Proportions")
plt.show()

#This pie chart shows the distribution of injury severity levels, providing insights into the proportion of fatal and non-fatal injuries.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a pivot table
crash_pivot = df.pivot_table(index="CRASH_DAY_OF_WEEK", columns="CRASH_HOUR", values="CRASH_RECORD_ID", aggfunc="count")

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(crash_pivot, cmap="coolwarm", annot=True, fmt=".0f")
plt.xlabel("Hour of the Day")
plt.ylabel("Day of the Week")
plt.title("Heatmap: Crash Occurrences by Day and Hour")
plt.show()

#This heatmap visualizes the frequency of crashes occurring at different hours of the day across each day of the week.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert CRASH_DATE to datetime format if it's not already
df["CRASH_DATE"] = pd.to_datetime(df["CRASH_DATE"])

# Extract month from CRASH_DATE
df["CRASH_MONTH"] = df["CRASH_DATE"].dt.month

# Count crashes per month
monthly_crash_counts = df["CRASH_MONTH"].value_counts().sort_index()

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(monthly_crash_counts.index, monthly_crash_counts.values, marker="o", linestyle="-", color="b")
plt.xlabel("Month")
plt.ylabel("Number of Crashes")
plt.title("Monthly Trend of Traffic Crashes")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid()
plt.show()

#This line plot shows the trend of crashes over months, helping identify seasonal crash patterns.
```

```python
# Selecting relevant columns
columns = ["POSTED_SPEED_LIMIT", "INJURIES_TOTAL", "LIGHTING_CONDITION", "WEATHER_CONDITION"]
df_selected = df[columns].dropna()

# Pair Plot
sns.pairplot(df_selected, hue="LIGHTING_CONDITION", diag_kind="kde")
plt.show()

#This plot visualizes relationships between speed limit, total injuries, lighting condition, and weather condition.
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(df["CRASH_HOUR"], fill=True, color="blue")
plt.title("Crash Density at Different Times of Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Density")
plt.xticks(range(0, 24))
plt.show()

#This plot shows the density distribution of crashes based on the hour of the crash.
```
